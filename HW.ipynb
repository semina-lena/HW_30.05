### Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ratings.csv')
ratings['date'] = pd.to_datetime(ratings['timestamp'], unit='s')
ratings.head()
```

```python
def groupby_function(data):
    return data.date.max() - data.date.min()
```

```python
df_time_life = ratings.groupby('userId').apply(groupby_function).reset_index()
df_time_life = df_time_life.rename(columns={ df_time_life.columns[1]: 'time_life' })
df_time_life.head()
```

```python
couts_ratings = ratings.userId.value_counts().reset_index()
couts_ratings = couts_ratings.rename(columns={'index': 'userId', 'userId': 'count_r'})
couts_ratings.head()
```

```python
df = pd.merge(couts_ratings, df_time_life, how = 'left')
df = df.loc[df.count_r >= 100]
df
```

```python
df['time_life'].mean()
```
